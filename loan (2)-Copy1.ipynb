```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyp
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
train_name = 'E:/python/train.csv'
test_name = 'E:/python/test.csv'

cat_train_features = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area', 'Loan_Status']
cat_test_features = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']

train = pd.read_csv(train_name,index_col=False)
train = train[cat_train_features]

test = pd.read_csv(test_name,index_col=False)
test = test[cat_test_features]

train
```

```python
le=LabelEncoder()

train = pd.DataFrame.dropna(train)
train.select_dtypes(include=[object]).columns.tolist()

test = pd.DataFrame.dropna(test)
test.select_dtypes(include=[object]).columns.tolist()

for x in train.select_dtypes(include=[object]).columns.tolist():
    train[x] = le.fit_transform(train[x])
    
for y in test.select_dtypes(include=[object]).columns.tolist():
    test[y] = le.fit_transform(test[y])

train.shape
```

```python
trainx = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']
trainx = train[trainx]

trainy = ['Loan_Status']
trainy = train[trainy]

testx = ['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']
testx = test[testx]

trainx
```

```python
lr = LinearRegression()

lr.fit(trainx,trainy)

y_pred = lr.predict(testx)

np.mean(y_pred)
```
